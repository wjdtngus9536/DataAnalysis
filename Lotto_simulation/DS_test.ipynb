```python
import random
import matplotlib.pyplot as plt

iteration_count = 10000
sample_counts = [50, 100, 200]
count_of_80 = {50:[], 100:[], 200:[]}

for sample_count in sample_counts:
    current_count_of_80 = []
    for _ in range(iteration_count):
        cnt = 0
        for i in range(sample_count):
            lotto_numbers = random.sample(range(1, 46), 3)
            if sum(lotto_numbers) == 80:
                cnt += 1
        if cnt > 100:
                cnt = 100
        current_count_of_80.append(cnt)
    count_of_80[sample_count] = current_count_of_80

plt.figure(figsize = (15, 5))
for i, sample_count in enumerate(sample_counts):
    plt.subplot(3, 1, i+1)
    plt.hist(count_of_80[sample_count], bins = range(101), align = 'left', edgecolor = 'black')
    plt.xlabel('count  of 80')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of number 80 in {sample_count}')
plt.tight_layout()
plt.show()
```

```python
#  6 sampleing without replacement
#   50, 100, 200 .
#   10000   1~6       


iteration_count = 10000
sample_counts = [50, 100, 200]
count_of_samples = {50:[], 100:[], 200:[]}

for sample_count in sample_counts:
    result = []
    for _ in range(iteration_count):
        lotto_numbers = random.sample(range(1, 46), 6)
        count_1_6 = random.choices(lotto_numbers, k = sample_count)
        cnt = 0
        for i in count_1_6:
            if 1 <= i <= 6:
                cnt += 1
        result.append(cnt)

    plt.figure(figsize = (15, 5))
    plt.hist(result, bins = range(17, 255), edgecolor = 'black')
    plt.xlabel('count')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {sample_count}')
```
