## 1. 로또 번호(1~45)를 무작위로 sampling with replacement로 50번, 100번, 1,000번 샘플링 하는 것을 50,000번 진행할 때 로또번호 숫자 중 짝수의 숫자가 나오는 횟수를 히스토그램으로 보이시오.

```python
import random
import matplotlib.pyplot as plt

iteration_count = 50000
sample_counts = [50, 100, 1000]
even_counts = {50:[], 100:[], 1000:[]}

for sample_count in sample_counts:
    current_even_counts = [] # 길이 5만 리스트(50개 중 짝수 개수)
    
    for _ in range(iteration_count):
        # 50개 추출을
        lotto_numbers = random.choices(range(1, 46), k=sample_count)
        even_count = sum(1 for num in lotto_numbers if num %2 == 0)
        current_even_counts.append(even_count)
        
    even_counts[sample_count] = current_even_counts
    
# 히스토그램 그리기

# 그림 크기 지정
plt.figure(figsize = (15, 5))

for i, sample_count in enumerate(sample_counts):
    plt.subplot(3, 1, i+1) # i+1 사용 이유 1이상부터 시작하기 때문에
    plt.hist(even_counts[sample_count], bins = range(sample_count), align='left', edgecolor = 'black')
plt.show()
```

## 2. 로또 번호에 대해 10만번 반복 동안 샘를링 횟수를 100번 500번 1000번 실시했을 때 숫자 23이 나오는 횟수를 기록하고, 나온 횟수를 0 ~ 100까지의 범위로 제한하여 히스토그램으로 표시하는 코드를 작성하시오

```python
import random
import matplotlib.pyplot as plt

iteration_count = 100000
sample_counts = [50, 100, 500, 1000]
count_of_23 = []

for sample_count in sample_counts:
    current_count_of_23 = []
    
    for _ in range(iteration_count):
        lotto_numbers = random.choices(range(1, 46), k = sample_count)
        current_count_of_23.append(lotto_numbers.count(23))
        
    # 나온 횟수를 0 ~ 100까지의 범위로 제한 100과 count중 작은 값 선택
    current_count_of_23 = [min(count, 100) for count in current_count_of_23]
    
    count_of_23.append(current_count_of_23)
    
plt.figure(figsize = (15, 5))

for i, sample_count in enumerate(sample_counts):
    plt.subplot(4, 1, i+1)
    plt.hist(count_of_23[i], bins = range(101), edgecolor = 'black')
    plt.xlabel('Number of Times 23 appeared')
    plt.ylabel('Frequency')
    plt.title('Histogram of the Number 23')
    
plt.tight_layout()
plt.show()
```

## 4. 로또 숫자 1~45까지 숫자 중 6개를 sampling without replacement로 샘플링한 뒤 숫자의 합을 구하고 그 작업을 1,000번, 10,000번, 100,000번 반복하는 시뮬레이션을 하고 결과를 그래프로 보여주는 파이썬 코드를 작성하시오

```python
import random
import matplotlib.pyplot as plt

simulation_counts = [1000, 10000, 100000]
results = []

for simulation_count in simulation_counts:
    current_results = []
    
    for _ in range(simulation_count):
        lotto_numbers = random.sample(range(1, 46), 6)
        current_results.append(sum(lotto_numbers))
    results.append(current_results)
    
plt.figure(figsize = (12, 8))

for i, simulation_count in enumerate(simulation_counts):
    plt.subplot(3, 1, i+1)
    plt.hist(results[i], bins = range(17, 255), edgecolor = 'black')
plt.show()
    
```

## 5. 로또 숫자의 합을 구하고 그 작업을 1000번, 10,000번, 100,000번 반복하고 이때 각각의 평균과 분산 값도 계산하여 결과를 보여주는 그래프를 작성하시오

```python
import random
import numpy as np
import matplotlib.pyplot as plt

simulation_counts = [1000, 10000, 100000]
results = []

for simulation_count in simulation_counts:
    current_results = []
    
    for _ in range(simulation_count):
        lotto_numbers = random.sample(range(1, 46), 6)
        current_results.append(sum(lotto_numbers))
    results.append(current_results)

    
plt.figure(figsize = (12, 8))

for i, simulation_count in enumerate(simulation_counts):
#     plt.subplot(3, 1, i+1)
    plt.hist(results[i], bins = range(17, 255), align = 'left', edgecolor = 'black')
    mean_value = np.mean(results[i])
    variance_value = np.var(results[i])
    plt.title(f'Histogram of the Sum of 6 Lotto Numbers ({simulation_count} Iterations)\nMean: {mean_value:.2f}, Variance: {variance_value:.2f}')
    plt.show()
```

## 6. 로또 숫자 합을 50번 100번, 200번씩 한다고 하자. 그 샘플링 작업을 10,000번 반복할 때 각(50, 100, 200번 샘플링)의 평균과 분산 값을 계산하여 결과를 그래프로 보여주는 파이썬 코드를 작성하시오

```python
import random
import numpy as np
import matplotlib.pyplot as plt

iteration_count = 10000

sample_counts = [50, 100, 200]

for sample_count in sample_counts:
    sums = []
    
    # 시뮬레이션 수행
    for _ in range(iteration_count):
        lotto_numbers = random.sample(range(1, 46), 6)
        total_sum = sum(lotto_numbers)
        sums.append(total_sum)
        
    # 평균과 분산 계산
    mean_value = np.mean(sums)
    variance_value = np.var(sums)
    plt.hist(sums, bins = range(7, 252), align = 'left')
    plt.xlabel('Sum of 6 Lotto Numbers')
    plt.ylabel('Frequency')
    plt.title(f'Historgram of the sum of 6 lotto numbers ({iteration_count} iterations, {sample_count} Samples Each)\nMean: {mean_value:.2f}, Variance: {variance_value:.2f}')
    plt.show()
    
```
